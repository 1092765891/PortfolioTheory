```python
portfolio_noEM = log_data[['AGG','SPY','QQQ','GLD']]
portfolio_siEM = log_data[['AGG','SPY','QQQ','EMB']]
noa_noEM = len(portfolio_noEM.columns)
noa_siEM = len(portfolio_siEM.columns)
```

```python
noa_noEM
```

### Portfolio without EM bonds

```python
Expected_Return_noEM = np.sum(portfolio_noEM.mean()* noEM)* semana
```

```python
Expected_Return_noEM
```

```python
Expected_Std_noEM = np.sqrt(np.dot(noEM.T,np.dot(portfolio_noEM.cov()*52,noEM)))
```

```python
Expected_Std_noEM
```

```python
Sharpe_noEM = Expected_Return_noEM / Expected_Std_noEM
```

```python
Sharpe_noEM
```

```python
portfolio_noEM.min()
```

### Portfolio with EM bonds

```python
Expected_Return_siEM = np.sum(portfolio_siEM.mean()* siEM)* semana
```

```python
Expected_Return_siEM
```

```python
Expected_Std_siEM = np.sqrt(np.dot(siEM.T,np.dot(portfolio_siEM.cov()*52,siEM)))
```

```python
Expected_Std_siEM
```

```python
Sharpe_siEM = Expected_Return_siEM / Expected_Std_siEM
```

```python
Sharpe_siEM
```

```python
portfolio_siEM.min()
```

### Simulations

```python
prets = []
pvols = []

for p in range(2500):
    weights = np.random.random(noa_noEM)
    weights /= np.sum(weights)
    prets.append(np.sum(portfolio_noEM.mean()* weights)* 252)
    pvols.append(np.sqrt(np.dot(weights.T, np.dot(portfolio_noEM.cov() * 252, weights))))
prets = np.array(prets)
pvols = np.array(pvols)

plt.figure(figsize = (10,8))
plt.scatter(pvols, prets, c = prets / pvols, marker = 'o')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label = 'Sharpe Ratio')
plt.title('Monte Carlo Simulation Efficient Frontier with no EM');
```

```python
prets = []
pvols = []

for p in range(2500):
    weights = np.random.random(noa_siEM)
    weights /= np.sum(weights)
    prets.append(np.sum(portfolio_siEM.mean()* weights)* 252)
    pvols.append(np.sqrt(np.dot(weights.T, np.dot(portfolio_siEM.cov() * 252, weights))))
prets = np.array(prets)
pvols = np.array(pvols)

plt.figure(figsize = (10,8))
plt.scatter(pvols, prets, c = prets / pvols, marker = 'o')
plt.grid(True)
plt.xlabel('expected volatility')
plt.ylabel('expected return')
plt.colorbar(label = 'Sharpe Ratio')
plt.title('Monte Carlo Simulation Efficient Frontier with EM');
```
